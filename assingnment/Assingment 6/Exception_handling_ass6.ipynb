# Task 1: Safe Division utility
## Write a program that:
### 1. Takes two inputs from the user: numerator and denominator
### 2. Uses try-except to handle:
###  ValueError(if input is not a number)
###  ZeroDivisionError(if denominator=0)
### 3. if no error occurs, print the result inside the else block
### 4. in the finally block,print:
###    "Operation Complete"

```python
numerator=int(input("Enter the numerator value: "))
denominator=int(input("Enter the denominator value: "))
try:
    result=numerator/denominator
except ZeroDivisionError as ex1:
    print(ex1)
except ValueError as ex2:
    print(ex2)
else:
    print("The Division result is: ",result)
finally:
    print("Operation Completed")       
```

# Task 2: Bill Calculator with Error Handling
### Given a list of products prices:
### prices =[120,350,'abc',500,-200,800]
### Write code that:
### 1. Iterates the list.
### 2. Tries to add only valid (positive numerical) prices to the total.
### 3. Handles:
###   TypeError if value is not a number
###   Custom exception using raise ValueError("Negative price not allowed")
### 4. Prints the running total.
###    Excepted behaviour:
### Skip inavlid items but continue processing 

```python
prices=[120,350,'abc',500,-200,800]
total =0
for price in prices:
    try: 
        if not isinstance(price,(int,float)):
            raise TypeError("It is not a number")
        
        if price<0:
            raise ValueError("Negative price not allowed")
        
        total=total+price
        print("Running Total: ",total)

    except TypeError:
        print(price,"is not a valid number")

    except ValueError as ex:
        print(ex)
        
print("Final Total: ",total)
```

# Task 3: Custom Exception: Age Validator
### 1. Write a function check_age(age) that:
###  Raises a custom exception ValueError ("Age must be between 1 and 120") if age is out of range.
###  2. In your main code:
###    Take age input from the user.
###    Use try-except to catch and print the custom error message.

```python
def check_age(age):
    if age<=1 or age>=120:
        raise ValueError("Age must be between 1 and 120")
try :
    age= int(input("Enter Age: "))
    check_age(age)
    print("Age is Valid")
except ValueError as ex:
    print(ex)
```

### 3. Handle:
###           FileNotFoundError

### 4. If successful, print first 3 lines of the file.

# Task 5: Mini Program: Safe Shopping Cart
### Create a program that:
### 1. Has a cart list --> cart =[]
### 2. Runs a loop asking the user to enter prices.
### 3. Stops when user enters 'q'
### 4. Inside the loop:
###   Convert input to a float
###   Handle ValueError if user
###   Raise custom exception if price is negative
### 5. At the end,print:
### Total items
### Total bill


```python
cart=[]
while True:
    price =input("Enter the price or Enter q to quit: ")
    if price=="q":
        break
    try:
        price=float(price)
        
        if price<0:
            raise ValueError("Negative Price is not allowed")
        
        cart.append(price)

    except ValueError as ex:
        print(ex)
print("Total Items: ",len(cart))
print("Total Bill",sum(cart))
```
