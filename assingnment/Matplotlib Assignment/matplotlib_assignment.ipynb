# Task 1: Line Plot (Sales Trend)
### Gather a Sales Dataset From Kaggle of your Choice and Then,
### 1. Create a line plot showing sales trend over months.
### 2. Add:
### • Title 
### • X-axis label
### • Y-axis label
### 3. Display the plot

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
sales=pd.read_csv('sales_data_sample.csv')
```

```python
sales.sample(10)
```

```python
month_order=['January','February','March','April','May','June','July','August','September','October','November','December']
```

```python
t=sales.groupby('Month')['numbers'].sum()
```

```python
t=t.reindex(month_order)
```

```python
plt.plot(t.index,t.values,marker='o')
plt.title('Sales Trends Over Months')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.xticks(rotation='vertical')
plt.show()
```

# Task 4: Multiple Bar Plot ()
### 1. Create a multiple bar chart comparing sales of Different Years
### 2. Use proper bar width and alignment.
### 3. Add legend, title, and axis labels

```python
df=pd.read_csv('mobile_sales_data.csv')
```

```python
df
```

```python
df['Inward Date']=pd.to_datetime(df['Inward Date'])
```

```python
df['Inward Date']=df['Inward Date'].dt.month_name()
```

```python
df
```

```python
df['Inward Date'].value_counts()
```

```python
m=df[df['Brand']=='Google']
d=df[df['Brand']=='Nokia']
p=df[df['Brand']=='Apple']
```

```python
m=m.groupby('Inward Date')['Price'].sum()
d=d.groupby('Inward Date')['Price'].sum()
p=p.groupby('Inward Date')['Price'].sum()
```

```python
m
```

```python
d
```

```python
p
```

```python
m.index=pd.CategoricalIndex(m.index,categories=['January','February','March','April','May','June','July','August','September','October','November','December'],ordered=True)
```

```python
d.index=pd.CategoricalIndex(d.index,categories=['January','February','March','April','May','June','July','August','September','October','November','December'],ordered=True)
```

```python
p.index=pd.CategoricalIndex(p.index,categories=['January','February','March','April','May','June','July','August','September','October','November','December'],ordered=True)
```

```python
m=m.sort_index()
d=d.sort_index()
p=p.sort_index()
```

```python
x=['January','February','March','April','May','June','July','August','September','October','November','December']
```

```python
plt.figure(figsize=(8,5))
plt.title('Brandwise Monthly Sales of Mobile Phones',color='red')
plt.xlabel('Month Name',color='red')
plt.ylabel('Total Sales',color='red')
plt.bar(np.arange(m.shape[0]) -0.2,m.values,color='green',label='Google',width=.2)
plt.bar(np.arange(d.shape[0]),d.values,color='blue',label='Nokia',width=.2)
plt.bar(np.arange(p.shape[0]) +0.2,p.values,color='red',label='Apple',width=.2)
plt.xticks(np.arange(m.shape[0]) -0.2,x,rotation='vertical')
plt.legend()
plt.show()
```
