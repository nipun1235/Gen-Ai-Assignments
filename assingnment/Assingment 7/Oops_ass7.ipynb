### (Extra) apply discount method

```python
class Product:
    
    def __init__(self,name,price,category):
        self.name=name
        self.price=price
        self.category=category
        
    
    def apply_discount(self,discount_percent):
        discounted_price=self.price-(self.price*discount_percent/100)
        return discounted_price
    
```

```python
class Product:
    
    def __init__(self,name,price,category):
        self.name=name
        self.price=price
        self.category=category

    def get_price(self):
        print(f"The Price of {self.name} is {self.__price}")
    
    def set_price(self,new_price):
        if new_price>0:
            self.__price=new_price
            print(f"The new price of {self.name} is {self.__price}")
```

# Task 3: Inheritance (Single-Level)

### Create a subclass ElectronicProduct that inherits from Product.
### Additional attribute:
### warrant_years
### Override the get_info() method to include warranty info.
### Create an object and demonstrate inheritance + overrinding.

```python
class Product:
    
    def __init__(self,name,price,category):
        self.name=name
        self.price=price
        self.category=category

class ElectronicProduct(Product):
    def __init__(self,name,price,category,warranty_years):
        super().__init__(name,price,category)
        self.warranty_years=warranty_years
    
    def get_info(self):
        print(f"Product name is {self.name} : {self.price}\n and it's category is {self.category}\n It has warranty of {self.warranty_years} year")
```

```python
product_1=ElectronicProduct("Laptop",85000,"Electronic Product",1)
```

```python
product_1.get_info()
```

# Task 5: Abstraction(Using Abstract Base Class)

### Create an abstract class Payment with abstract method:
### process_payment(amount)
### Then create two subclasses :
### CreditCardPayment
### UPI Payment
### Both Overrides process_payment() with simple print statements.

```python
from abc import ABC,abstractmethod

class Payment(ABC):
    @abstractmethod
    def Process_payment(self,amount):
        pass

class CreditCardPayment(Payment):
    def Process_payment(self, amount):
        self.amount=amount
        print(f"The Credit Card Payable amount is {amount}")

class UPIPayment(Payment):
    def Process_payment(self, amount):
        self.amount=amount
        print(f"The UPI payment amount is {amount}")
```

```python
CC=CreditCardPayment()
```

```python
Upi=UPIPayment()
```

```python
CC.Process_payment(17650)
```

```python
Upi.Process_payment(18378)
```

# Task 6: Magic Methods & Operator Overloading

### Add the following to your Product class:
### 1. __str__ method
### Returns a readable string:
### Product(name,prace,category)
### 2. Operator Overloading(__add__)
### Allow:
### product1 + product2
### To return the total combined price.
### Test this with two product objects.

```python
class Product:

    def __init__(self,name,price,category):
        self.name=name
        self.price=price
        self.category=category
    
    def __add__(self,other):
        return Product(self.name + " & " + other.name,self.price + other.price,self.category + " & " + other.category)

    def __str__(self):
        return f"{self.name},{self.price},{self.category}"
    
product_1=Product("PC",45000,"Computer")
product_2=Product("CPU",50000,"Operator")

Combined_price=product_1+product_2
print(Combined_price)
```

# Task 7: Mini Project: Simple Inventory System(OOP Only)

### Create two classes:
### Class: Inventory
### Attributes:
### products --> list to store product objects
### Methods:
### add_product(product)
### remove_product(name)
### get_total_value() --> sums prices of all products
### show_all_products() --> prints info for each product
### Class: Store
### Attributes:
### store_name
### inventory --> all inventory object
### Methods:
### add_new_product()
### show_summary() --> prints total items and value|

```python
class Product:

    def __init__(self, name, price, category):
        self.name = name
        self.price = price
        self.category = category

    def get_info(self):
        print(f"Name: {self.name}")
        print(f"Price: {self.price}")
        print(f"Category: {self.category}")

    def __str__(self):
        return f"Product({self.name}, {self.price}, {self.category})"

    def __add__(self, other):
        return self.price + other.price


class Inventory:

    def __init__(self):
        self.products = []

    def add_product(self, product):
        self.products.append(product)

    def remove_product(self, name):

        for product in self.products:

            if product.name == name:
                self.products.remove(product)
                print("Product removed successfully")
                return

        print("Product not found")

    def get_total_value(self):

        total = 0

        for product in self.products:
            total += product.price

        return total

    def show_all_products(self):

        for product in self.products:
            print(product)


class Store:

    def __init__(self, store_name):
        self.store_name = store_name
        self.inventory = Inventory()

    def add_new_product(self):

        name = input("Enter product name: ")
        price = float(input("Enter product price: "))
        category = input("Enter category: ")

        product = Product(name, price, category)

        self.inventory.add_product(product)

    def show_summary(self):

        print("\nStore Name:", self.store_name)
        print("Total Items:", len(self.inventory.products))
        print("Total Value:", self.inventory.get_total_value())

store = Store("Nipun Store")

p1 = Product("Laptop", 50000, "Electronics")
p2 = Product("Mobile", 20000, "Electronics")
p3 = Product("Headphones", 3000, "Accessories")

store.inventory.add_product(p1)
store.inventory.add_product(p2)
store.inventory.add_product(p3)

print("All Products:")
store.inventory.show_all_products()

print("\nSummary:")
store.show_summary()

print("\nCombined Price:")
print(p1 + p2)

print("\nRemoving Mobile...")
store.inventory.remove_product("Mobile")

print("\nUpdated Products:")
store.inventory.show_all_products()

print("\nUpdated Summary:")
store.show_summary()
```
