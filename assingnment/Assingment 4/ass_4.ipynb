# Task 7: Mini Product - Export Discounted Prices 

### Create a dictionary:
### prices ={
###    "Mouse": 500
###    "Keyboard": 800
###    "Monitor": 7000
###    "Pendrive": 400
###    "Camera": 5000
### }
### Ask the user for a discount percentage.
### Write discounted prices into discount_report.txt using:
### Product|Original Price| Discounted price
### After Writing,read the file and print it to the terminal.
### Extra : Write a sumamry at the bottom of the file:
### Total Items: X
### Average Discounted Price: Y

```python
prices ={"Mouse":500, "keyboard":800, "Monitor":7000, "Pendrive":400, "Camera":5000}

Discount_percent=int(input("Enter Discount Percentage: "))

file=open('discount_report.txt','w')

for product in prices:
    original_price= prices[product]
    discounted_price= original_price-(original_price*Discount_percent/100)

    file.write(product + "|"+ str(original_price)+ "|" + str(discounted_price) +"\n")
file.close()

file=open('discount_report.txt','r')
print(file.read())
file.close()
```
