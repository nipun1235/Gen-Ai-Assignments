# Task 2- Recursive Function: Factorial Utility

## Create a recursive function factorial(n) that:
### 1. Return the factorial n.

```python
def factorial(n):
    if n==1 or n==0:
        return 1
    else :
        return n*factorial(n-1)
```

```python
factorial(10)
```

### 2. Handles two edge Cases : n==1 and n==0

```python
factorial(0)
```

```python
factorial(1)
```

### 3. Prints an error message if n is negative

```python
def factorial(n):
    if n==1 or n==0:
        return 1
    elif  n<0:
        print("Error Input Entered")
    else:
        return n*factorial(n-1)
```

```python
## test with:
# factorial(5)
factorial(5)
```

```python
# factorial(0):
factorial(0)
```

```python
# test with factorial(-3):
factorial(-3)
```

## Task - 3 Lambda Function: GST Calculator
### Create a lambda function gst that returns price after adding 18% GST.
### gst=lambda price: price +(0.18*price)
### print(gst(100))# should return 118

```python
gst=lambda price:price+(price*0.18)
```

```python
print(gst(100))
```

### Extra(optional): Create another to compute final price after gst+discount together

```python
def apply_discount(price,discount_percent):
   gst=lambda price:price+(price*0.18)
   gst_price=gst(price)
   final_price=gst_price-(gst_price*discount_percent/100)
   return final_price
```

```python
print(apply_discount(3000,16))
```

```python
def average_price(price_list):
    avg=sum(price_list)/(len(price_list))
    return avg
```

```python
price_list=[500, 1000, 50, 680, 985]
print(average_price(price_list))
```

```python
def get_max_price(price_list):
    max_price=max(price_list)
    return max_price
```
