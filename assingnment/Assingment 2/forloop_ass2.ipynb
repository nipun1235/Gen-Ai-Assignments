# Task 2: Process Multiple Orders(for loop)

### 1. Given a list of order amount: order = [1200,2500,800,1750,3000], use a for loop to apply the discount rules from Task 1 to each order and print a summary table showing : order_amount -> discount% -> final_amount.

```python
orders= [1200,2500,800,1750,3000]
for order_amount in orders:

    if order_amount>=2000 :

        discount = 15/100

        final_amount= order_amount-(order_amount*discount)

    elif 2000>order_amount>=1500 :

        discount = 10/100

        final_amount=order_amount-(order_amount*discount)

    elif 1500>order_amount>=1000 :

        discount= 7/100

        final_amount=order_amount-(order_amount*discount)

    else :
        final_amount=order_amount       
    print("order amount:",order_amount , "discount :",discount , "final amount:",final_amount )  
     
```

### 2. Also compute and print the total revenue after discounts.

```python
orders= [1200,2500,800,1750,3000]
final_revenue=0
for order_amount in orders:

    if order_amount>=2000 :

        discount = 15/100

        final_amount= order_amount-(order_amount*discount)

    elif 2000>order_amount>=1500 :

        discount = 10/100

        final_amount=order_amount-(order_amount*discount)

    elif 1500>order_amount>=1000 :

        discount= 7/100 

        final_amount=order_amount-(order_amount*discount)

    else :
        
        final_amount=order_amount       
    
    print("order amount:",order_amount , "discount :",discount , "final amount:",final_amount )
    final_revenue=final_revenue+final_amount
    print("final revenue is:",final_revenue)  
     
```
