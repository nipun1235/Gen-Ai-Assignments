# Task 5 : Loop Control with conditions (break & continue)

### 1. Given a list of daily sales: daily=[200,150,0,400,50,-1,300], iterate through it with a for loop and
### -> if a day's value is -1,treat it as currupted data and break the loop(stop processing)
### -> if a day's value is 0,treat it as a day with no sales and continue(skip adding to revenue).
### -> For valid positive sales, add to total_sales and print the running total.

```python
daily=[200,150,0,400,50,-1,300]
total_sales=0

for day_value in daily:

    if day_value==-1:
     
     print("currupted data")
     break

    elif day_value==0:

        print("no sale on this day")
        continue

    if day_value >0:
     
     total_sales=total_sales+day_value
     print("the running total is: ",total_sales)
```

### 2. Print the final total after loop completes (or stop due to currupted data).

```python
daily=[200,150,0,400,50,-1,300]
total_sales=0


for day_value in daily:

    if day_value==-1:
     
     print("currupted data")
     break

    elif day_value==0:

        print("no sale on this day")
        continue

    if day_value >0:
     
     total_sales=total_sales+day_value
     print("the running total is: ",total_sales)
print("the total sales is: ",total_sales)
```
