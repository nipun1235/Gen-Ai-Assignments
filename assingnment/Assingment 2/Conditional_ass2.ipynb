# Task 1: Discount Rules(if/elif/else)

### 1. Write a program that reads an integer order_amount from the user using input().

```python
order_amount=int(input("Enter your Order amount:"))
print("Your Order amount is:",order_amount)
```

### 2. Apply the following  discount rules and print the final amount:
### order_amount>=2000--->15% discount
### 1500<= order_amount< 2000--->10% discount
### 1000<=order_amount<1500---> 7% discount
### otherwise--->0% discount

```python
order_amount=int(input("Enter the Order amount"))
if order_amount>=2000:

    print(order_amount-(order_amount*15/100))

elif 1500<=order_amount<2000:

    print(order_amount-(order_amount*10/100))

elif 1000<=order_amount<1500:

    print(order_amount-(order_amount*7/100))
    
else :
    print(order_amount)
```

### 3. Ensure you convert input to a number and handle the case when the input is non-numeric by displaying an error message and exiting the program.

```python
order_amount=(input("Enter the order amount: "))

if order_amount.isdigit():
    order_amount=int(order_amount)
    print(order_amount)
else :
    print("Invalid Input")
```
