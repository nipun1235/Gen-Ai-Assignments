# User Menu (While loop + break/continue)

### 1. Implement a while-loop driven menu that keeps asking the user to choose an action until they press q to quit.
### Menu options: 
### 1-->Add order amount to a running list
### 2-->Show all orders and totals after applying discounts
### q--> Quit

```python
orders = []

while True:

    print("1 - Add Order")
    print("2 - Show Orders")
    print("q - Quit")

    choice = input("Enter your choice: ")

    if choice == "1":
        amount = int(input("Enter order amount: "))
        orders.append(amount)

    elif choice == "2":

        for order_amount in orders:

            if order_amount >= 2000:
                final_amount = order_amount - (order_amount * 15 / 100)

            elif 2000>order_amount >= 1500:
                final_amount = order_amount - (order_amount * 10 / 100)

            elif 1500>order_amount >= 1000:
                final_amount = order_amount - (order_amount * 7 / 100)

            else:
                final_amount = order_amount

            print("Order amount is :", order_amount,"Final Amount after discount is :", final_amount)

    elif choice == "q":
        break
```
