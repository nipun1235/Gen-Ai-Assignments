# Task 9: Mini Use Case: Sales Data Analysis
### Create a DataFrame:
### sales =
### 'Day': ['Mon', 'Tue", 'Wed', 'Thu', 'Fri'],
### 'Revenue': [1200, 1500, 900, 2000, 1800]
### Perform:
### 1. Total revenue
### 2. Average daily revenue
### 3. Day with highest revenue
### 4. Days where revenue > average
### 5. Plots Revenue vs Day

```python
sales ={
'Day': ['Mon', 'Tue', 'Wed', 'Thu', 'Fri'],
'Revenue': [1200, 1500, 900, 2000, 1800]
}

df= pd.DataFrame(sales)
```

```python
df
```

```python
df['Revenue'].sum()
```

```python
avg=df['Revenue'].mean()
```

```python
avg
```

```python
df.loc[df['Revenue'].idxmax(),'Day']
```

```python
df[df['Revenue'] > avg]
```

```python
df.plot(x='Day',y='Revenue',kind='bar')
```
